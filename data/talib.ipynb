```python
#추가한 지표
def MACD(data):
    data['ema_short_12']=data['Close'].ewm(12).mean()
    data['ema_long_26']=data['Close'].ewm(26).mean()
    data['MACD']=data['ema_short_12']-data['ema_long_26']
    return data


def TRIX(data,n=12):
    ema_n=data['Close'].ewm(n).mean()
    ema_n1=data['Close'].ewm(n-1).mean()
    data['TRIX']=(3*ema_n-3*ema_n1)/(3*ema_n1)
    return data


def adx(data,n=14):
    data['adx']=ta.trend.ADXIndicator(data['High'],data['Low'],data['Close'],window=n).adx()
    return data

def stochastic_ocil(data,n_1=5,n_2=14):
    stoch = ta.momentum.StochasticOscillator(data['High'], data['Low'], data['Close'], window=n_1, smooth_window=n_2)
    data['%K'] = stoch.stoch()
    data['%D'] = stoch.stoch_signal()
    return data



def aroon_oscillator(data, n=20):
    aroon = ta.trend.AroonIndicator(data['High'], data['Low'], window=n)
    data['Aroon_Oscillator'] = aroon.aroon_indicator()
    return data



def momentum(data,n=10):
    data['Momentum']=data['Close'].diff(n)
    return data

def calculate_awesome_oscillator(data, short_period=5, long_period=20):
    data['Short_MA'] = data['Close'].rolling(window=short_period).mean()
    data['Long_MA'] = data['Close'].rolling(window=long_period).mean()
    data['Awesome_Oscillator'] = data['Short_MA'] - data['Long_MA']
    return data


def calculate_accelerator_oscillator(data, period=5):
    data['AC'] = (data['High'].rolling(window=period).mean() - (data['Close'] + data['Low']) / 2).rolling(window=period).mean()
    return data


def calculate_balance_of_power(data):
    data['BOP'] = ((data['Close'] - data['Open']) / (data['High'] - data['Low'])) * data['Volume']
    return data

```

```python
data=MACD(df)
data=TRIX(data)
data=adx(data)
data=stochastic_ocil(data)
data=aroon_oscillator(data)
data=momentum(data)
data=calculate_awesome_oscillator(data)
data=calculate_accelerator_oscillator(data)
data=calculate_balance_of_power(data)
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
hc1 = linkage(ss_index_df.T, 'ward')
dendrogram(hc1,show_leaf_counts=True,labels=ss_index_df.columns)
# x축 변수값 세로로 출력
plt.xticks(rotation='vertical')
plt.show()
```

```python
cut_line = 150
clusters = fcluster(hc1, cut_line, criterion='distance')

# 군집별로 열을 출력
for cluster_num in range(1, max(clusters) + 1):
    cluster_columns = ss_index_df.columns[clusters == cluster_num]
    print(f'군집{cluster_num}: {", ".join(cluster_columns)}')
```

```python
###순열중요도 계산 알고리즘###(논문 참고)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
from sklearn.metrics import mean_absolute_percentage_error as mape

# 1. 데이터 준비
# 군집 내 변수들 선택
cluster_features = ss_index_df[['rsi', 'cci', 'williumR', 'plusdi']]

target_variable=df['volatility'].iloc[30:]
# 2. 모델 학습
X_train, X_test, y_train, y_test = train_test_split(cluster_features, target_variable, test_size=0.2, random_state=42)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
# 회귀 모델 평가
y_pred = model.predict(X_test)
res=mape(y_test, y_pred)
print(res)

# 3. 순열 중요도 계산
perm_importance = permutation_importance(model, X_test, y_test, n_repeats=30, random_state=42)

# 4. 결과 해석
# 중요도 결과 출력
for i, feature in enumerate(cluster_features.columns):
    print(f"{feature}: {perm_importance['importances_mean'][i]}")
```
